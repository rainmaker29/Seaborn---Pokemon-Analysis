```python
import pandas as pd
import seaborn as sb
```

```python
import seaborn as sns
```

### Now that we've imported seaborn and pandas,we'll load our dataframe and plot various plots on it


```python
df = pd.read_csv('pokemon.csv')
```

```python
df.head()
```

```python
df = df.rename(columns = {"Generation":"Stage"})
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

## Scatter Plot

```python
%matplotlib inline
sns.lmplot(x='Attack',y='Defense',data=df)
plt.figure(figsize=(25,10))
```

### Removing the regression line

```python
sns.lmplot(x='Attack',y='Defense',data=df,fit_reg=False,hue='Stage')
```

# Boxplot

```python
sns.boxplot(data=df)
```

# Dropping few unnecessary columns for plotting

```python
stats_df = df.drop(['Legendary','Total','Stage'],axis=1)
```

```python
sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')

sns.violinplot(x='Type 1',y='Attack',data=df)
```

```python
#Appropriate Colors list
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.violinplot(x='Type 1',y='Attack',data=df,palette=pkmn_type_colors)
```

```python
sns.swarmplot(x='Type 1',y='Attack',data=df,palette=pkmn_type_colors)
```

# Overlaying the swarm and violin plot

```python
plt.figure(figsize=(16,6))

sns.violinplot(x='Type 1',y='Attack',data=df,inner=None,palette=pkmn_type_colors)

sns.swarmplot(x='Type 1',y='Attack',data=df,color='k',alpha=0.7)
```

# Melting several stats 

```python
melted = pd.melt(stats_df,id_vars=["Name","Type 1", "Type 2"],var_name="Stat")
melted.head()
```

```python
sns.swarmplot(x='Stat',y='value',data=melted,hue='Type 1')
```

```python
#Customizations
```

```python
plt.figure(figsize=(20,10))
sns.swarmplot(x='Stat',y='value',data=melted,hue='Type 1',split=True,palette=pkmn_type_colors)
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

# Various plots 

```python
#Heatmap is plotted for correlatioin among attributes
```

```python
corr = stats_df.corr()

sns.heatmap(corr)
```

```python
sns.distplot(df.Attack)
```

```python
sns.catplot(x='Type 1',y='Attack',data=df,hue='Stage',col='Stage',kind='swarm')
```

```python
sns.jointplot(x='Attack',y='Defense',data=df)
```
